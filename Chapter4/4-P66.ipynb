```python
import matplotlib.pyplot as plt
from pandas import DataFrame
plt.figure()
income=[1.47,1.62,1.78,1.94,2.38,2.60]
plt.plot(income,color='black',linewidth=2,linestyle=':',marker='s',label='Income')
plt.annotate('Largest!',xy=(5,2.60),xytext=(3.1,2.54),arrowprops=dict(arrowstyle='->'))
plt.title('Income chart')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Income(RMB Ten Thousand)')
plt.xlim(0,5)
plt.ylim(0.0,3.0)
plt.xticks(range(0,6),('2012','2013','2014','2015','2016','2017'))
plt.legend(loc='upper right')
plt.grid()
plt.savefig('Income chart.jpg',dpi=400,bbox_inches='tight')
```

```python
from pandas import Series
import matplotlib.pyplot as plt
data=Series([1.47,1.62,1.78,1.94,2.38,2.60],
            index=['2012','2013','2014','2015','2016','2017'])
fig=plt.figure(figsize=(6,6))
#plt.subplots_adjust( )

ax1=fig.add_subplot(2,2,1)
data.plot(kind='line',xticks=[],yticks=[1.50,1.75,2.00,2.25,2.50],ax=ax1,color='blue')
#data.plot(kind='line',title='line chart',yticks[1.50,1.75,2.00,2.25,2.50])
plt.title('line chart')
plt.xlabel('Year',fontsize=12)
plt.ylabel('Income',fontsize=12)
#plt.yticks[1.50,1.75,2.00,2.25,2.50]
plt.xticks(range(0,6,2),('2012','2014','2016'))
plt.subplots_adjust(bottom=2,top=3)
ax1.plot(data)


ax2=fig.add_subplot(2,2,2)
data.plot(kind='box',fontsize='small',xticks=[],yticks=[1.50,1.75,2.00,2.25,2.50],ax=ax2)
plt.title('box-whisker plot')
plt.xlabel('2012-2017',fontsize=12)
plt.ylabel('Income',fontsize=12)
plt.subplots_adjust(wspace=0.3)



ax3=fig.add_subplot(2,1,2)
data.plot(kind='bar',use_index=True,fontsize='small',ax=ax3,label='Income')
plt.title('bar chart')
plt.ylabel('Income',fontsize=12)
plt.legend(loc='upper left')

```
